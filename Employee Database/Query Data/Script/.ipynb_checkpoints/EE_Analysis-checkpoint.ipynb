```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pyodbc
import psycopg2

# SQL Alchemy
from sqlalchemy import create_engine

# Import database connection string containing username, password and hostname
from config import password
```

```python
# Connect to database
engine = create_engine(f'postgresql://postgres:{password}@127.0.0.1:5433/EEDatabase')
conn = engine.connect()
```

```python
# Import salaries into Pandas
Salary = pd.read_sql('Select * From "Salaries"',conn)
Salary.head()
```

```python
# Create histogram to visualize salary ranges
plt.hist(Salary.salary)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```
